<b> Safety based on Marital Status</b> <br>


<font color='red'> The finding is that married folks involved in collission represent a higher percentage. This does not support the notion that married folks are safer drivers. </font> <br>
Count all Collision claims and store in a variable.

```python
c= df[(df['Claim Reason']=="Collision")]['Claim Reason'].count()
```

Find how many of each of the marital status filed a claim for collision and store in variables.

```python
x = df[(df['Marital Status']=="Married") & (df['Claim Reason']=="Collision")]['Marital Status'].count()
y = df[(df['Marital Status']=="Divorced") & (df['Claim Reason']=="Collision")]['Marital Status'].count()
z = df[(df['Marital Status']=="Single") & (df['Claim Reason']=="Collision")]['Marital Status'].count()
```

Out of all claims for collision, find percent of married and single. <br>


```python
print('Married as percent of total collisions is ' + str(round((x/c*100),2)) + '%')
print('Divorced as percent of total collisions is ' + str(round((y/c*100),2)) + '%')
print('Single as percent of total collisions is ' + str(round((z/c*100),2))+ '%')
```
